```python
import pandas as pd
import sys
```

```python

dframes=[]

for p in sys.argv[1:]:
    dframes.append(pd.read_csv(p))
```

```python
for df in dframes:
    df.columns=['id','marks']
    df.drop_duplicates(subset=['id'], keep='last',inplace=True)
    df.reset_index(inplace=True)
    df.drop("index",axis=1,inplace=True)
    df['marks'].fillna(value=0,inplace=True)
```

```python
mydf=[]

for j in range(0,len(dframes)):
    for i in range(0,len(dframes[j])):
        try:
             dframes[j]['marks'][i]=pd.to_numeric( dframes[j]['marks'][i])
        except: 
             mydf.append([dframes[j].loc[i,'id'],dframes[j].loc[i,'marks'],"File" + str(j+1)])
             dframes[j].drop(i,inplace=True)        
```

```python
mydf=pd.DataFrame(mydf)
mydf.columns=['id','marks','Filename']
mydf.to_csv("log.csv",mode='a',header=True,index=False)
# myfile=pd.DataFrame(myfile)
# mydf=pd.concat([mydf,myfile],axis=1)
```
